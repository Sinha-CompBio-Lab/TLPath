```python
telomere_df_fie =  #your telomere file location
UNI_features_dir = "./tlpath/UNI_features/"
output_dir = "./TLPath_results"
```

```python
telomere_df = pd.read_csv(telomere_df_fie, sep=",")
telomere_dict  = create_telomere_dict(telomere_df)
tissues_data = process_all_tissues(UNI_features_dir, telomere_dict)
datasets = build_dataset(tissues_data, telomere_dict)
```

## Train TLPath

```python
config = TLPathConfig()

tlpath = TLPath(
    config=config,
    telomere_file=telomere_df_fie,
    UNI_features_dir=UNI_features_dir,
    output_dir=output_dir
)

results = tlpath.run_pipeline()
save_result_summary(results, output_dir)    
```
