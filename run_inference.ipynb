```python
import pandas as pd

import numpy as np
from tlpath.utils import (
    predict_telomere_length,
    predict_all_tissues,
    predict_batch,
)
```

### Define Model directory

```python
# Tissue models directory
models_dir = '../TLPath/models/' 
```

### Read your dataset

```python
# Read Pancreas features
pancreas_df = pd.read_csv('example_data/Pancreas.csv')
pancreas_feature = np.fromstring(pancreas_df['features'].iloc[0], sep=',')

# Read Lung features
lung_df = pd.read_csv('example_data/Lung.csv')
lung_feature = np.fromstring(lung_df['features'].iloc[0], sep=',')

# Read Prostate features
prostate_df = pd.read_csv('example_data/Prostate.csv')
prostate_feature = np.fromstring(prostate_df['features'].iloc[0], sep=',')

# Verify the shape and contents
print("Pancreas feature shape:", pancreas_feature.shape)
print("Lung feature shape:", lung_feature.shape)
print("Prostate feature shape:", prostate_feature.shape)

# Optional: Confirm the first few values match the original
print("\nFirst 5 values of Pancreas feature:")
print(pancreas_feature[:5])
print("\nFirst 5 values of Lung feature:")
print(lung_feature[:5])
print("\nFirst 5 values of Prostate feature:")
print(prostate_feature[:5])
```

### Run inference

```python
# Single tissue prediction
telomere_length_pancreas = predict_telomere_length(models_dir, pancreas_feature, "Pancreas")
```

```python
print(f'Predicted Telomere length for pancreatic Sample is {telomere_length_pancreas}')
```

```python
# Multiple tissue prediction
tissue_features = {
    "Pancreas": pancreas_feature,
    "Lung": lung_feature,
    "Prostate": prostate_feature
}
all_predictions = predict_all_tissues(models_dir, tissue_features)
```
